```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Sample Healthcare Data ---
data = {
    'Department': ['Cardiology', 'Orthopedics', 'Neurology', 'Pediatrics', 'Cardiology',
                   'Orthopedics', 'Neurology', 'Pediatrics', 'Cardiology', 'Orthopedics'],
    'Treatment_Cost': [50000, 32000, 42000, 15000, 48000, 30000, 41000, 18000, 51000, 29000],
    'Stay_Days': [6, 4, 7, 3, 5, 4, 6, 2, 6, 3]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# --- 1. Bar Chart: Average Treatment Cost per Department ---
avg_cost = df.groupby('Department')['Treatment_Cost'].mean()
plt.figure(figsize=(7,4))
avg_cost.plot(kind='bar', color='teal')
plt.title("Average Treatment Cost per Department")
plt.xlabel("Department")
plt.ylabel("Average Cost (INR)")
plt.grid(axis='y')
plt.show()

# --- 2. Scatter Plot: Stay Duration vs Treatment Cost ---
plt.figure(figsize=(6,4))
plt.scatter(df['Stay_Days'], df['Treatment_Cost'], color='orange', edgecolor='black')
plt.title("Relationship between Stay Duration and Treatment Cost")
plt.xlabel("Hospital Stay (Days)")
plt.ylabel("Treatment Cost (INR)")
plt.grid(True)
plt.show()
```
